```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_tesla = pd.read_csv('TSLA.csv',date_parser='Date')
df_tesla
```

```python
df_tesla = df_tesla.set_index('Date')
```

```python
df_tesla
```

```python
df_tesla['Open'].plot(figsize=(15,6))
```

```python
df_tesla['Opne:10 days rolling'] = df_tesla['Open'].rolling(window=10,min_periods=1).mean()
```

```python
df_tesla.head(15)
```

```python
df_tesla[['Open','Opne:10 days rolling']].plot(figsize=(15,6))
```

```python
#  Expanding 
df_tesla['Open'].expanding().mean().plot(figsize=(15,6))
```
